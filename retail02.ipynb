```python
import pandas as pd
import numpy as np
```

```python
df0 = pd.read_csv('sales_store.csv')
```

```python
df0.head(10)
```

```python
df0.info()
```

```python
df1 = df0.copy()
```

```python
# Date to datetime
df1['Date'] = pd.to_datetime(df1['Date'], format='%d/%m/%Y')
```

```python
df1.info()
```

```python
df1['IsHoliday'].replace({True: 1, False: 0}, inplace=True)
```

```python
df1.head(10)
```

```python
df1['year'] = pd.DatetimeIndex(df1['Date']).year
```

```python
df1
```

```python
df_2010 = df1[df1['year'] == 2010]
df_2010.reset_index(inplace=True)
df_2010
```

```python
df_2011 = df1[df1['year'] == 2011]
df_2011.reset_index(inplace=True)
df_2011
```

```python
df_2012 = df1[df1['year'] == 2012]
df_2012.reset_index(inplace=True)
df_2012
```

```python
# exportar dataset por year

df_2010.to_csv('data_2010.csv', index=False)
df_2011.to_csv('data_2011.csv', index=False)
df_2012.to_csv('data_2012.csv', index=False)
```
